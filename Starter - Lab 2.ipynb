# Lab 1

```python
# Load the data
data = pd.read_csv("department-sales.csv", index_col= 0, parse_dates=True)
data.head()
```

```python
# Change date format to yyyy-mm-dd
data.index = pd.to_datetime(data.index, format="%d/%m/%Y")
data.head()
```

```python
# Set index frequency to weekly
data.index.freq = "W-FRI"
```

# Lab 2 - Visualization

* Plot weekly sales using the Weekly_Sales column and add a title to the chart

* Highlight holiday weeks with vertical dashed red lines

```python
plt.figure(figsize=(10,4))
plt.plot(data.Weekly_Sales)

for date, row in data.iterrows():
  if row['IsHoliday']:
    plt.axvline(date, color = 'red', linestyle = "--", lw = 0.5)

plt.title("Weekly Sales")
plt.show()
```
